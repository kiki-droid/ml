```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import datasets, metrics, linear_model
from sklearn.model_selection import train_test_split

# load the boston dataset
boston = datasets.load_boston()
print(boston.DESCR)
  
# defining feature matrix(X) and response vector(Y)
X = boston.data
print(X,"\n")
print(X.shape,"\n")

Y = boston.target
print(Y,"\n")
print(Y.shape,"\n")
  
# splitting X and y into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=3)
  
# create linear regression object
reg = linear_model.LinearRegression()
  
# train the model using the training sets
reg.fit(X_train, Y_train)

print("# Training data\n") 
print(X_train.shape,"= X_train\n")
print(Y_train.shape,"= Y_train\n")
print("# Testing data\n")
print(X_test.shape,"= X_test\n")
print(Y_test.shape,"= Y_test\n")

# regression coefficients
print('Intercept: ', reg.intercept_)
print('Coefficients: ', reg.coef_)
print("\n")

Y_train_predict = reg.predict(X_train)
rmse = (np.sqrt(metrics.mean_squared_error(Y_train, Y_train_predict)))
r2 = round(reg.score(X_train, Y_train),2)

print("The model performance for training set")
print("--------------------------------------")
print('Root Mean Squared Error: {}'.format(rmse))
print('MSE: {}'.format(r2))
print("\n")

Y_test_pred = reg.predict(X_test)
rmse = (np.sqrt(metrics.mean_squared_error(Y_test, Y_test_pred)))
r2 = round(reg.score(X_test, Y_test),2)

print("The model performance for test set")
print("--------------------------------------")
print("Root Mean Squared Error: {}".format(rmse))
print("MSE: {}".format(r2))
print("\n")

plt.scatter(Y_test, Y_test_pred)
plt.xlabel('Test Y')
plt.ylabel('Predicted Y')
plt.show()
```
