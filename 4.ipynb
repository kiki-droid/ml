1.2. From the variable, extract the numerical information into a new
variable, say scores.

```python
scores = pd.read_csv('/content/wine.csv')
print(scores)
```

1.3. Calculate the row-wise,column-wise, and overall mean, median,
mode, and standard deviation of the dataset. 

```python
print("Mean:Computed column-wise:");
meanData = scores.mean();
print(meanData);
```

```python
print("Median:Computed column-wise:");
medianData = scores.median();
print(medianData);
```

```python
print("Mode:Computed column-wise:");
modeData = scores.mode();
print(modeData);
```

```python
print("Standard deviation:Computed column-wise:");
stdData = scores.std();
print(stdData);
```

```python
print("Mean:Computed row-wise:");
meanData = scores.mean(axis=1);
print(meanData);
```

```python
print("Median:Computed row-wise:");
medianData = scores.median(axis=1);
print(medianData);
```

```python
from scipy.stats import mode
print("Mode:Computed row-wise:");
modeData = mode(scores, axis=1).mode;
print(modeData);
```

```python
print("Standard deviation:Computed row-wise:");
stdData = scores.std(axis=1);
print(stdData);
```

```python
Table_mean = np.nanmean(scores.values)
Table_median = np.nanmedian(scores.values)
Table_std = np.nanstd(scores.values)
print("Table mean:")
print(Table_mean)
print("Table median:")
print(Table_median)
print("Table std:")
print(Table_std)
```

```python
df = np.array(scores)
vals,counts = np.unique(df,return_counts=True)
index = np.argmax(counts)
Table_mode = vals[index]
print("Table mode:")
print(Table_mode)
listOfUniqueValues = zip(vals,counts)
print('Unique Values along with occurrence Count')
# Iterate over the zip object
for elem in listOfUniqueValues:
   print(elem[0] , ' Occurs : ' , elem[1], ' times')
```

2.2. Take any 5 features from the ‘wine.csv’ datasets and load them in
a matrix. Thereafter, create the covariance matrix of the same.

```python
df2 = scores.sample(n = 5, axis = 'columns')
print(df2)
```

```python
mat = df2.to_numpy()
print(mat)
```

```python
print("Covariance matrix:")
print(np.cov(mat))
```
