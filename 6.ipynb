```python
import numpy as np
import matplotlib.pyplot as plt
import random
X = [random.uniform(1,3) for i in range(6)] 
y = [1.7+0.5*x for x in X]

X = np.asarray(X)
m = 0
c = 0
l = 0.001

n = float(len(X))
iterations = 3

for i in range(iterations):
    y_pred = m*X + c
    cost = sum((y-y_pred)**2)
    print(cost)
    D_m = -(2/n) * (sum((y - y_pred) * X))
    D_c = -(2/n) * (sum(y - y_pred))
    m = m - l*D_m
    c = c - l*D_c
    y_pred = m*X + c
    plt.scatter(X,y)
    plt.plot(X,y_pred)
```
